```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(".\\photon_gas_mc\\Photon_gas.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Calculated Average as a solid line
plt.plot(data["BetaEpsilon"], data["Calc_Avg_Val"], '-', label="Calculated Average", linewidth=2, color='tab:blue')

# Theoretical Average as points
plt.plot(data["BetaEpsilon"], data["Theo_Avg_Val"], 'o', label="Theoretical Average", markersize=5, alpha=0.8, color='tab:orange')

# Labels and title
plt.xlabel(r"$\beta \, \epsilon$", fontsize=12)
plt.ylabel("Average Value", fontsize=12)
plt.title("Comparison of Calculated vs Theoretical Averages", fontsize=14, fontweight='bold')

# Grid and layout
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Optional: zoom in on small values for clarity
plt.ylim(bottom=0)  

plt.show()
```
