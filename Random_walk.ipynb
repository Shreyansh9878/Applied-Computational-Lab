```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the files
results_1 = pd.read_csv("./Random_walk_res/10000-50/results.dat", 
                        delim_whitespace=True, header=None)
results_2 = pd.read_csv("./Random_walk_res/10000-100/results.dat", 
                        delim_whitespace=True, header=None)
results_3 = pd.read_csv("./Random_walk_res/10000-200/results.dat", 
                        delim_whitespace=True, header=None)

# Plot both
plt.figure(figsize=(7,5))

plt.plot(results_1[0], results_1[1], label="10000-50")
plt.plot(results_2[0], results_2[1], label="10000-100")
plt.plot(results_3[0], results_3[1], label="10000-200")

plt.xlabel("Column 1")
plt.ylabel("Values")
plt.title("Visualization of results.dat (Two Runs)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
rms_1 = pd.read_csv("./Random_walk_res/10000-50/rms.dat", delim_whitespace=True, header=None)
rms_2 = pd.read_csv("./Random_walk_res/10000-100/rms.dat", delim_whitespace=True, header=None)
rms_3 = pd.read_csv("./Random_walk_res/10000-200/rms.dat", delim_whitespace=True, header=None)

plt.figure(figsize=(7,5))
plt.plot(rms_3[0], rms_3[1], linestyle='-.', label='10000-200')
plt.plot(rms_2[0], rms_2[1], linestyle='--', label='10000-100')
plt.plot(rms_1[0], rms_1[1], linestyle='-', label='10000-50')
plt.xlabel("Column 1")
plt.ylabel("Column 2")
plt.title("Visualization of rms.dat")
plt.grid(True)
plt.legend() 
plt.show()
```
