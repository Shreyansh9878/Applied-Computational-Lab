```python
import numpy as np
```

```python
r = 0.5
a = 2*r
iteration = 50000

total_pts = iteration*iteration
pts_in_circle = 0

for i in np.linspace(-r,r,iteration):
    for j in np.linspace(-r,r,iteration):
        if (i**2) + (j**2) <= (r**2):
            pts_in_circle+=1

print(4*pts_in_circle/total_pts)
```

```python
print((np.pi - (4*pts_in_circle/total_pts))*100/np.pi)
```
